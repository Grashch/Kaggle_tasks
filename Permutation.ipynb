```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import eli5
from eli5.sklearn import PermutationImportance
```

```python
data = pd.read_csv('/content/FIFA 2018 Statistics.csv')
data.head()
```

```python
y = (data['Man of the Match'] == 'Yes')
features = [i for i in data.columns if data[i].dtype == 'int64']
X = data[features]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
model = RandomForestClassifier(n_estimators=100, random_state=1).fit(X_train, y_train)
```

```python
perm = PermutationImportance(model, random_state=1).fit(X_test, y_test)
```

```python
# eli5.show_weights(perm, feature_names=X_test.columns.tolist()) - было так
eli5.show_weights(perm, feature_names=features)
```
